```python
"""
This file is a temp file, which combines all the data that we have into a unified dataset.
Important is to contain sent, claim, row_index, column_index, formula
"""
```

```python
import pandas as pd
import os
```

```python
CLAIMS_EXTRACTED_BY_US_DIR = "data/claims_extracted_by_us/"
```

```python
OUT_FILE = "data/main_annotated_dataset_12-16-2019.csv"
```

```python
def get_files_from_dir(basepath):
    files = []
    for entry in os.listdir(basepath):
        if os.path.isfile(os.path.join(basepath, entry)):
            files.append(os.path.join(basepath, entry))
    return files
```

```python
us_files = get_files_from_dir(CLAIMS_EXTRACTED_BY_US_DIR)
```

```python
us_files
```

```python
us_df = pd.DataFrame()
```

```python
for file in us_files:
    print(file)
    temp_df = pd.read_csv(file)
    temp_df["file"] = file
    # remove unwanted rows
    us_df = pd.concat([us_df, temp_df])
```

```python
us_df = us_df.rename(columns={"Text": "sent", "Published value": "claim", "Fixed_Formulas": "formula",
                       "Look-up value": "row_index", "Look-up year": "column", "Scenario": "scenario"})
```

```python
us_df = us_df[["sent", "claim", "formula", "row_index", "column", "scenario", "file"]]
```

```python
us_df
```

```python
# main_df = pd.concat((us_df, chapters_df))
main_df = us_df
```

```python
main_df
```

```python
len(main_df.drop_duplicates(subset=["sent", "claim", "formula", "row_index", "column"]))
```

```python
main_df = main_df.drop_duplicates(subset=["sent", "claim", "formula", "row_index", "column"]).reset_index(drop=True)
```

```python
main_df
```

```python
main_df.to_csv(OUT_FILE, index=False)
```
