```python
import json
import matplotlib.pyplot as plt
```

```python
stats_dict = {"column": [56, 10, 9, 4, 2, 19], "row_index": [25, 7, 3, 1, 2, 62], "all_props": [0, 0, 0, 0, 0, 0], "template_formula": [30, 9, 2, 1, 2, 56], "file": [55, 10, 5, 1, 3, 26]}
```

```python
stats_dict
```

```python
def get_pies_prop(prop, stats_dict):
    d = stats_dict[prop]
    labels = ["Ver_num_{}".format(i+1) for i in range(len(d)-1)]
    labels.append("Der")
    sizes = [i for i in d]
    # Plot
    fig = plt.gcf()
    fig.set_size_inches(12,12)
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
    plt.axis('equal')
    plt.show()
```

```python
get_pies_prop("file", stats_dict)
```
