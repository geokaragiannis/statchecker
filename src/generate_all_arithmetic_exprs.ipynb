```python
import itertools
import math
import operator
```

```python
cols = ["A", "B", "C", "D", "E"]
ops = ["+", "-", "*", "/"]
```

```python
n = len(cols)
```

```python
def get_all_perms(cols):
    ret_list = []
    # no singles
    for k in range(2, len(cols)):
        ret_list += itertools.permutations(cols, k)
    ret_list += list(itertools.permutations(cols))
    return ret_list
```

```python
def get_ops_permutations(ops, k):
    return list(itertools.product(ops, repeat=k))
```

```python
cols = ["A", "B", "C", "D", "E"]
```

```python
exprs_list = []
for col_num in range(2, len(cols)+1):
    perms = itertools.product(cols, repeat=col_num)
    ops_permutations = get_ops_permutations(ops, col_num-1)
    for perm in perms:
        for op_comb in ops_permutations:
            k = ""
            for j in range(0, len(perm)-1):
                k += perm[j] + op_comb[j]
                if j == len(perm) - 2:
                    k += perm[j+1]
            exprs_list.append(k)
```
